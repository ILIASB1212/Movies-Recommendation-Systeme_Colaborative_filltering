```python
import pandas as pd
```

```python
movie=pd.read_csv("movies_small.csv")
rate=pd.read_csv("ratings.csv")
```

```python
import re
def movies(name):
    name=name.split("(")[0]
    return name.lower()
movie["title"]=movie["title"].apply(movies)
```

```python
movie.head(2)
```

```python
data=rate.pivot(index="movieId",columns="userId",values="rating").fillna(0)
```

```python
data
```

```python
movie_rated=rate.groupby("userId")["rating"].count()
user_rate=rate.groupby("movieId")["rating"].count()
```

```python
data=data.loc[user_rate[user_rate>10].index]
data=data.loc[:,movie_rated[movie_rated>50].index]
```

```python
data
```

```python
from scipy.sparse import csr_matrix
data.reset_index(inplace=True)
sparce=csr_matrix(data)
print(sparce)
```

```python
from sklearn.neighbors import NearestNeighbors
knn=NearestNeighbors(metric="cosine",algorithm="brute",n_neighbors=20)
knn.fit(sparce)
```

```python
from difflib import get_close_matches
def app(name):
    try:
        list_of_movies=movie["title"].to_list()
        matching=get_close_matches(name,list_of_movies,n=1, cutoff=0.6)[0]
        
        movie_id = movie[movie["title"] == matching]["movieId"].values[0]
        pivot_index = data[data["movieId"] == movie_id].index[0]
            
        distance,indice=knn.kneighbors(sparce[pivot_index],n_neighbors=11)
        zipped=zip(indice.squeeze().tolist(), distance.squeeze().tolist())    
        sorted_movie=sorted(zipped,key=lambda x:x[1],reverse=False)[1:]
        
        lista=[]
        for idc,dist in sorted_movie:
            x=data.iloc[idc]["movieId"]
            title = movie[movie["movieId"] == x]["title"].values[0]
            lista.append(title)
        return lista
    except:
        return f'{name} was not found'
    
    
    
```

```python
app("hacker")
```
